```python
# Toh upar wale rate se 1969 ka growth tha as compared to 1968
```

```python

data=df_pr.values
```

```python
gdp_change=[0]
for i in range(1,len(data)):
    prev=data[i-1][3]
    curr=data[i][3]
    gdp_change.append(round(((curr-prev)/prev)*100,2))
    
```

```python
df_pr.assign(GDP=gdp_change)
```
